```python
import random
def populate_list(length):
    lst = []
    for i in range(length):
        lst.append(random.randint(10, 50))
    return lst
```

```python
def sum_list(lst):
    sum = 0
    for num in lst:
        sum += num
    return sum
```

```python
user_input = int(input("Please enter an integer between 5 and 15: "))
```

```python
if user_input < 5 or user_input > 15:
    print("Invalid input!")
else:
    lst = populate_list(user_input)
    print("The elements of the arrays are:", lst)
    sum_of_list = sum_list(lst)
    print("The sum is:", sum_of_list)
```
